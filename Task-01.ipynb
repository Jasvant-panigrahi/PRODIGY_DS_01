```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\Prodigy\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_6548260.csv")
df.head()
```

```python
gender_counts=df["Region"].value_counts()
bar_width=0.9
x=range(len(gender_counts.index))
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel("Region")
plt.ylabel("Count")
plt.title("Distribution of the region")
plt.xticks(x, gender_counts.index, rotation=45)
plt.tight_layout()
plt.show()
```
